```python
# program to process student marks and determine pass/fail status
def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    average = total / len(marks)
    
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

student_marks = [60, 45, 70, 50, 55]
average, result = process_result(student_marks)
print("Average Marks:", average)
print("Result:", result)
```
