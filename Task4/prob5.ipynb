```python
# program to check student attendance and determine eligibility for exams
def check_attendance(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    
    if total_days == 0:
        return 0, "No Attendance Data"
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
percentage, result = check_attendance(attendance)
print(f"Attendance Percentage: {percentage:.2f}%")
print("Eligibility Status:", result)
```
