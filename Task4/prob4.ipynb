```python
# program to calculate cab fare based on distance and peak hour status
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if peak_hour:
        fare += fare * 0.25   # 25% extra
    return fare

while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()
        
        if peak_input not in ["yes", "no"]:
            print("Please enter 'yes' or 'no' for peak hour.")
            continue
        
        peak_hour = True if peak_input == "yes" else False
        total_fare = calculate_fare(distance, peak_hour)
        print(f"Total Cab Fare: ₹{total_fare:.2f}")
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break
            
    except ValueError:
        print("Please enter valid numeric distance.\n")
```
