```python
# program to validate mobile recharge amount
def validate_recharge(amount, valid_plans):
    if amount < 50:     # Check minimum amount
        print(" Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:     # Check if amount is in valid plans
        print(" Invalid recharge plan.")
        print("Available plans are:", valid_plans)
        return False
    
    return True

valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(amount, valid_plans):
            print(" Recharge successful!")
            break
        else:
            print(" Please try again.\n")
            
    except ValueError:
        print(" Please enter a valid numeric amount.\n")
```
