```python
# program to manage a smart parking system
def smart_parking_system(capacity, vehicle_logs):
    current_parked=0
    peak_usage=0
    for log in vehicle_logs:
        if log=="IN":
            current_parked+=1
        elif log=="OUT":
            if current_parked>0:
                current_parked-=1

        if current_parked>peak_usage:
            peak_usage=current_parked
    print("Currently Parked Vehicles:",current_parked)
    print("Peak Parking Usage:",peak_usage)
    if current_parked>capacity:
        print("Parking Status: FULL - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
smart_parking_system(capacity,vehicle_logs)
```
