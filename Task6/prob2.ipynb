```python
#program to estimate delivery time based on distance, traffic level, and weather conditions
def estimate_delivery_time(distance,traffic_level,weather):
    base_speed=30
    base_time=(distance/base_speed)*60

    if traffic_level=="Low":
        traffic_delay=5
    elif traffic_level=="Medium":
        traffic_delay=10
    elif traffic_level=="High":
        traffic_delay=20
    else:
        traffic_delay=0
    
    if weather=="Clear":
        weather_delay=0
    elif weather=="Cloudy":
        weather_delay=5
    elif weather=="Rainy":
        weather_delay=15
    elif weather=="Storm":
        weather_delay=25
    else:
        weather_delay=0

    final_time=base_time+traffic_delay+weather_delay
    print("Estimated Delivery Time:",round(final_time),"minutes")

estimate_delivery_time(8,"High","Rainy")
```
